```python
import pandas as pd
import numpy as np
import ecopy as ep
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
OTU_matrix = pd.read_csv("taxonomic.bacteria.csv", index_col=0)
meta = pd.read_table("Metadata_with_Host.tsv", index_col=0)
```

```python
#Transform the data with hellinger
OTU_hell = ep.transform(OTU_matrix, method="hellinger")
```

```python
OTU_hell.head()
```

```python
shannon_H = pd.DataFrame(ep.diversity(OTU_hell, method='shannon', breakNA=True, num_equiv=True))
```

```python
shannon_H.columns = ["Shannon_Index"]
```
